#CALCULO PI

a partir del ejemplo en libro Classic Computers Science Problems de David Kopec
https://github.com/davecom

donde presenta una solucion del tipo secuencia para calcular el numero Pi

`ej: 4/1 - 4/3 + 4/5 - 4/7 + 4/9 - 4/11 + .... - 4/1000000`

dando como resultado al millon de iteraciones: **3.14159**


se realiza propuesta de mejora al calculo

dando como resultado a las 29 iteraciones y 18 promedios: **3.1415926535897**

```python
#@title PI Original
# ORIG  https://github.com/davecom/ClassicComputerScienceProblemsInPython/blob/master/Chapter1/calculating_pi.py
def calculate_pi(n_terms: int) -> float:
    numerator: float = 4.0
    denominator: float = 1.0
    operation: float = 1.0
    pi: float = 0.0
    for _ in range(n_terms):
        pi += operation * (numerator / denominator)
        denominator += 2.0
        operation *= -1.0
    return pi

#if __name__ == "__main__":
print(calculate_pi(1000000))
print(calculate_pi(100))
```

```python
#@title PROPUESTA
def calculate_pi(n_terms: int) -> float:
    numerator: float = 4.0
    denominator: float = 1.0
    operation: float = 1.0
    pi: float = 0.0

    pi_ant1: float = 4
    pi_ant2: float = 0
    pi_prom1: float = 0
    pi_prom2: float = 0

    for _ in range(n_terms):
        pi_ant2 = pi_ant1
        pi_ant1 = pi
        pi += operation * (numerator / denominator)
        denominator += 2.0
        operation *= -1.0

    pi_prom1 = (pi + pi_ant1) /2
    pi_prom2 = (pi_ant1 + pi_ant2) /2
    pi = (pi_prom1 + pi_prom2) /2

    return pi

#if __name__ == "__main__":
#print(calculate_pi(350)) # PAR:POR ABAJO  IMPAR:POR ARIBA

pi1 = calculate_pi(34) # 100
pi2 = calculate_pi(35) # 101
print(pi1)
print(pi2)
print( (pi1 + pi2) /2 )
```

calcula Pi en una secuencia con promedios en cascada
reduce el numero de iteraciones y aumenta la precisión rápidamente
aporte realizado por Antonio Molina

Como:
-Calcula el promedio para los N últimos números de la secuencia
-a continuación calcula los promedios de los resultados
-y luego itera al siguiente nivel realizando calculando el promedio hasta que solo queda un numero

results:
iter, prom
  9,   6   err:5e-5    3.14153820036173
 11,   7   err:6e-6    3.1415986833943497
 29,   18  err:9e-15   3.1415926535897833

```python
#@title version final para https://github.com/davecom/ClassicComputerScienceProblemsInPython/blob/master/Chapter1/calculating_pi.py

# calcula Pi en una secuencia con promedios en cascada
# reduce el numero de iteraciones y aumenta la precision rapidamente
# aporte realizado por Antonio Molina
def calculate_pi(n_iter: int, n_prom = 1) -> float:
    numerator: float = 4.0
    denominator: float = 1.0
    operation: float = 1.0
    pi: float = 0.0

    dato = []
    prom = []

    #calcula secuencia
    for i in range(n_iter):
        pi += operation * (numerator / denominator)
        denominator += 2.0
        operation *= -1.0
        #acumula ultimos N resultados para promediar
        if(i>=n_iter-n_prom): dato.append(pi)

    #calcula promedios en cascada
    while(len(dato) > 1):
      prom = []
      for i, num in enumerate(dato[0:-1]):
          prom.append( (num + dato[i+1]) /2 )
      dato = prom.copy()

    return dato[0]

if __name__ == "__main__":
    print(calculate_pi(n_iter=29, n_prom=18))

#bulk test
#Pi = 3.141592653589793
#for i in range(1, 30):
#  for N in range(1, i+1):
#    P = calculate_pi(i, N)
#    print(i, N, P, Pi-P)
#print("Pi  ", Pi)

#results
#iter, prom   error      result
#  9,   6   err:5e-5    3.14153820036173
# 11,   7   err:6e-6    3.1415986833943497
# 29,   18  err:9e-15   3.1415926535897833
```
